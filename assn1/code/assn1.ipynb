```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
def computeSequence(a, b, x, n):
    sz_a = np.size(a)
    sz_x = np.size(x)
    if(n == sz_x+1):
        _sum = 0
        for i in range(sz_a):
            _sum += a[i]*x[sz_x-1-i]
        _sum += b
        return np.append(x,_sum)
    elif(n > sz_x+1):
        new_x = computeSequence(a,b,x,n-1)
        return computeSequence(a,b,new_x,n)
    elif (n <= sz_x):
        return x[:n]
    
def computeActual(n):
    return np.array([(4**(-k))/3 for k in range(n)])
```

```python
a = np.array([2.25, -0.5])
x = np.array([1/3, 1/12])
```

```python
t = computeSequence(a,0,x,80)
```

```python
plt.figure(figsize=(16,10))
plt.semilogy(range(80),computeSequence(a,0,x,80), base=10, label='Computed Solution')
plt.semilogy(range(80), computeActual(80), base=10, label='Exact Solution')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Semilog plot')
# plt.savefig('./plot.png', format='png')
plt.show()
```
